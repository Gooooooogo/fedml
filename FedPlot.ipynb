```python
import matplotlib.pyplot as plt
import pandas as pd

# 读取Excel文件
xls_file = pd.ExcelFile('result_N_T.xlsx')

# 获取所有sheet的名称
sheet_names = xls_file.sheet_names

# 初始化存储第二列和第五列数据的字典
sheet_data = {}
names =['Fastslowmo',
        'Nag',
       'Avg',
       'Mo',
       'Slowmo']
colors=['black','red','blue','green','purple']
for sheet_name in sheet_names:
    # 读取特定sheet的数据
    df = pd.read_excel(xls_file, sheet_name=sheet_name)
    filtered_df = df[df.iloc[:, 4] > 80]
    # 获取第二列和第五列的数据
    column2_data = filtered_df.iloc[:, 1].tolist()  # 第二列
    column4_data = filtered_df.iloc[:, 3].tolist()  # 第四列
    column5_data = filtered_df.iloc[:, 4].tolist()
    # 存储数据
    sheet_data[sheet_name] = {
        'column2': column2_data,
        'column4': column4_data,
        'column5': column5_data
    }

N=[2,4,8,16]
T=[10,20,40,80,160]
plt.figure(figsize=(40,30))
for t in range(len(T)):
    for n in range(len(N)):
        plt.subplot(5, 4, t*4+n+1 )
        #plt.figure(figsize=(2,2))  # 设置图形大小
        for name, color in zip(names,colors):
            plt.plot(sheet_data[name+'_'+str(N[n])+'_'+str(T[t])]['column2'], sheet_data[name+'_'+str(N[n])+'_'+str(T[t])]['column5'], label=name, color=color, linewidth=1)
        plt.title('N: '+str(N[n])+' T: '+ str(T[t]))  # 设置标题
        plt.xlabel('Total iteration')  # 设置X轴标签
        plt.ylabel('acc')  # 设置Y轴标签
        plt.legend()  # 显示图例
        plt.grid(True)  # 显示网格线
plt.show()  # 显示图形


```
